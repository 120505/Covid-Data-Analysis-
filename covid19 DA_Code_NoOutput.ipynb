### Task 3.3: find maximum infection rate for all of the countries. 

```python
covid_countries_max=[covid_countries.loc[country].diff().max() for country in covid_countries.index]
covid_countries_max
```

```python
len(covid_countries_max)
```

```python
covid_countries.head()
```

```python
covid_countries["max_infection_rate"]=covid_countries_max
covid_countries.head()
```

### Task 3.4: create a new dataframe with only needed column 

```python
covid_max_rate=pd.DataFrame(covid_countries["max_infection_rate"])
```

### Types of Join: 4
    Inner Join: Intersection of Rows
    Outer Join: Union of Rows
    Left Join: Left Dataset + Intersection
    Right Join: Right Dataset + Intersection

```python
final_data=covid_max_rate.join(happiness,how="inner")
final_data.head()
```

```python
final_data.shape
```
